# Section 2 : Next, developing equations ...

```python
x1 = 1
y1 = 7
x2 = 2
y2 = 10

#The slope is m
#M = difference value between the points. How much the line moves 
m = (y2 - y1) / (x2 - x1) 

#The y intercept is "b"
#That is where we begin on the y-axis where the x is zero
b = y1 - m*x1

#The full equation
print("y = ", m, "x + ", b)
```

# Section 5 : Graph the solution

The number of people afflicted with the common cold in the winter months dropped steadily by 50 each year since 2004 until 2010. In 2004, 875 were inflicted.

Find the linear function that models the number of people afflicted with the common cold C as a function of the year, t. When will no one be afflicted?

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy
from sympy import symbols
from sympy.solvers import solve

x1 = 0 #2004
y1 = 875 
x2 = 1 #2010
y2 = 825 #-50

#Develop the equation y = mx + b
m = (y2 - y1) / (x2 - x1) 
b = y1 - m*x1
print("y = ", m, "x + ", b)
print(-50.0*5 +  875.0)

#For the graph
xmin = 0 # start year
xmax = 18 #end year
ymin = 0 #start point
ymax = 900 #end point

#For the line on the graph
y3 = m * xmin + b
y4 = m * xmax + b

#Basic setup for the graph
fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

#Add details to the graph
ax.set_xlabel("time")
ax.set_ylabel("cases")
ax.grid(True)

#ax.set_xticks(np.arange(xmin, xmax, 2)) #set the tick mark on x-axis
#ax.set_yticks(np.arange(ymin, ymax, 2)) #tick marks side values of the graph on the axis

#Plot the linear function as a red line
plt.plot([xmin, xmax], [y3, y4], 'r')

plt.show()

x = symbols('x')

#Put the equation here
eq = 875 - 50*x
solution = solve(eq, x)
print("x = ", solution[0])
```
