# Section 1 : Simplest way to solve a system of equations. Set each equation equal to zero

```python
from sympy import * #symbolic math library

x,y = symbols('x y')

#Equations set equal to zero
first = 2*x + y - 1
second = x - 2*y + 7

#The solution
print(linsolve([first, second], (x, y)))
```

# Section 2 : Nicer looking output

```python
from sympy import *

x, y = symbols('x y')

first = 2*x + y - 1
second = x - 2*y + 7

# Parse finite set answer as coordinate pair
solution = linsolve([first, second], (x,y))
x_solution = solution.args[0][0] #Behind the scenes it is a 2 dimensional array
y_solution = solution.args[0][1]

#Print a coordinate pair
print("(", x_solution, ",", y_solution, ")")
```

# Section 3 : Factor to change from "0=" to "y=" preparing to graph

```python
from sympy import *
from sympy.plotting import plot

var('x y')

#First equation set equal to zero, ready to solve
first = 2*x + y - 1

#Second equation set equal to zero, ready to solve
second = x - 2*y + 7

#Solve
solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]
print("Solution = (", x_solution, ",", y_solution, ")")

#Sympy syntax for equation equal to zero, ready to factor
y_first = Eq(first, 0) #Equation set equal to zero

#Sympy solve for y
y1 = solve(y_first,y)

#Same two steps for second equation
y_second = Eq(second, 0)
y2 = solve(y_second, y)

#Show factored results
print("y = ",y1[0])
print("y = ",y2[0])

#Plot solution
x = symbols('x')
plot(y1[0], y2[0], (x,-10,10))
```
