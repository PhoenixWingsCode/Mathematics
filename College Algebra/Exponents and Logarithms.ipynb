# Section 4 : 

A good use of natural log : how long will it take for an investment to double in value?
Start with the continuous growth formula:
A = Pe**rt

If the amount is double, then annunity is two times the principle:
2P = Pe**rt

Divide both the sides by P and get this:
2 = e**rt

Because of the base e, take the natural log of both the sides and get this : 
In(2) = rt

Then divide by r to solve for t.
In(2)/r = t

```python
import math

#Rate of increase as a decimal
r = .02

#Natural log of 2 divided by rate
t = round(math.log(2)/r)

print("It will double in ", t, " years")
```

# Section 5 : Graphing

```python
import matplotlib.pyplot as plt
import numpy as np
import math

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 10*(xmax-xmin)
x = np.linspace(xmin, xmax, points)
x2 = np.linspace(.01, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [ymin, ymax], 'b') # blue y axis

# Line 1
# y1 = 2**x
y1 = math.e**x
plt.plot(x, y1)

# Line 2
# y2 = np.log2(x2)

y2 = np.log(x2)
plt.plot(x2, y2)

# Line 3
y3 = x
plt.plot(x, y3)

plt.show()
```

# Section 10 : Convert to scientific calculation

```python
import math

a = .0005

exp = math.floor(math.log(a, 10))
n = round(a/(10**exp),2)

# This addresses rounding errors
if n>=10:
    exp = exp + 1
    n = a/(10**exp)

print(a, " = ", n, "**10**", exp)
```
