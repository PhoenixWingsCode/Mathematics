# Section 1 : 

The word quad means square
Any time you graph a quadratic its a parabola
The general form of quadratic equation is y = Ax(2) + Bx + C
The point at which the parabola turns that is symmetrical that is vertex. x = -B/2A
The roots are = -B/2A +- sqrt(B(2) - 4AC)/2A

Solve a quadratic equation

```python
#Coefficient values
import math
a = 1
b = 5
c = 6

print("y = ", a, "x**2 + ", b, "x + ", c)

#Vertex
vx = -b / (2*a)
vy = a*(vx**2) + b*vx + c
print("Vertex : (", vx, "," , vy, ")")

#Roots
d = b**2 - 4*a*c
if d>=0:
    root_1 = (-b + math.sqrt(d))/(2*a)
    root_2 = (-b - math.sqrt(d))/(2*a)
    print("Roots: x = ", root_1, " and x = ", root_2)
else:
    print("No real roots")
```

# Section 2 : Graph a quadratic function

```python
import matplotlib.pyplot as plt
import numpy as np
import math

a = 2
b = 4
c = 1

print("y = ", a, "x**2 + ", b, "x + ", c)

#Vertex
vx = -b / (2*a)
vy = a*(vx**2) + b*vx + c
print("Vertex : (", vx, "," , vy, ")")

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 10 * (xmax -  xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

#Graph the parabola
y1 = a*x**2 + b*x + c
plt.plot(x, y1)

#Plot the vertex point
plt.plot([vx], [vy], 'ro')

#Find and plot the Roots
d = b**2 - 4*a*c
if d>=0:
    root_1 = (-b + math.sqrt(d))/(2*a)
    root_2 = (-b - math.sqrt(d))/(2*a)
    plt.plot([root_1, root_2], [0,0], 'go')
    print("Roots: x = ", root_1, " and x = ", root_2)
else:
    print("No real roots")
```

# Section 4 : Sliders to show how a, b, and c affect the graph

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

# All graphing happens in this function
def f(a, b, c):
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10
    points = 10 * (xmax - xmin)
    x = np.linspace(xmin, xmax, int(points))

    y = a * x**2 + b * x + c

    plt.figure(figsize=(6, 4))
    plt.axis([xmin, xmax, ymin, ymax])
    plt.plot([xmin, xmax], [0, 0], 'b')  # x-axis
    plt.plot([0, 0], [ymin, ymax], 'b')  # y-axis
    plt.plot(x, y, label=f'y = {a}x² + {b}x + {c}')
    plt.title(f'y = {a}x² + {b}x + {c}')
    plt.grid(True)
    plt.legend()
    plt.show()

# Create interactive sliders
interactive_plot = interactive(f, a=(1, 9), b=(-9, 9), c=(-9, 9))
display(interactive_plot)
```
