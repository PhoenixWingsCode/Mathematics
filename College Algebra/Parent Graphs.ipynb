# Section 1 : Graphs with no slope and no x value
y = 5

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 4*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax]) #window size
plt.plot([xmin, xmax], [0,0], 'b') #blue x axis
plt.plot([0, 0], [ymin, ymax], 'b') #blue y axis

plt.plot([xmin, xmax], [5, 5], 'r')

plt.show()
```

# Section 10 : Square root graphs

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 4 * (xmax - xmin)

x = np.linspace(0, xmax, int(points))  # Only non-negative x values

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])  # window size
plt.plot([xmin, xmax], [0, 0], 'b')  # blue x axis
plt.plot([0, 0], [ymin, ymax], 'b')  # blue y axis

plt.plot(x, np.sqrt(x), label='y = √x')
#plt.plot(x, -np.sqrt(x), label='y = -√x')

plt.title('Square Root Graphs')
plt.grid(False)
plt.legend()
plt.show()
```

# Section 16 : Sliders

```python
# sliders and zoom
# y = a

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

#Define the graphing function
def f(a, zoom):
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    points = 10*xmax-xmin
    x = np.linspace(xmin, xmax, points)

    plt.axis([xmin, xmax, ymin, ymax])  # window size
    plt.plot([xmin, xmax], [0, 0], 'b')  # blue x axis
    plt.plot([0, 0], [ymin, ymax], 'b')  # blue y axis

    #Line 1
    y1 = a
    y2 = a
    plt.plot([xmin, xmax], [y1, y2])
    plt.show()

#Set up the sliders
interactive_plot = interactive(f, a=(-9, 9), zoom=(1, 100))
interactive_plot
```
