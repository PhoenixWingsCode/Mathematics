# Section 4 : Display axis lines

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()

plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax], [0, 0], 'b') # blue x axis xmin -> 0 xmax -> 0 b represents blue
plt.plot([0,0], [ymin, ymax], 'b') # blue y axis

plt.show()
```

# Section 6 : Plot several points as a function

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()

plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

for x in range(10):
    y = 0.5*x + 1
    plt.plot([x], [y], 'ro')

plt.show()
```

# Section 7 : Graph and table of (x,y) values

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()

plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

print("x \t y")
for x in range(-5,10):
    y = 0.5*x + 1
    plt.plot([x], [y], 'ro')
    print(x, "\t", y)

plt.show()
```

# Section 8 : Use an array as inputs

```python
import matplotlib.pyplot as plt
#! pip install numpy
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points) #This function helps to create an array where to (start, end, how many points)
#This saves the time of having a loop
#An array of x values

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

y = 2*x + 1 #Here y becomes an array of y values
plt.plot(x, y, 'r') # r -> red line RGB and CMYK work well for other colors complete spelling pink, 
plt.show()
```

# Section 9 : More ways to customize a graph

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Some Graph")
ax.grid(True)

ax.set_xticks(np.arange(xmin, xmax, 2)) #set the tick mark on x-axis
#arange is differnt kind of array (start, end, how often the points to appear)
ax.set_yticks(np.arange(ymin, ymax, 2)) #tick marks side values of the graph on the axis

y=2*x + 1
plt.plot(x, y, label='y=2x+1')
plt.plot([4], [6], 'ro', label='point')
plt.plot(x, 3*x, label='steeper line') #here (x, y, label)

plt.legend() #The legend label to show the labels specified above
plt.show()
```
