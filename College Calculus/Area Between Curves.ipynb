```python
import sympy
from sympy import symbols, Eq

x, y, c = symbols('x y c')

# graph
eq1 = x**2
eq2 = x
sympy.plot(eq1, eq2, (x, -0.5, 1.5))
```

# Section 2 : Find intersections

```python
# find the intersection points

xPoints = sympy.solve(eq1, eq2)
xPoints
```

```python
# find the intersection points

xPoints = sympy.solve(Eq(eq1, eq2))
xPoints
```

# Section 3 : Subtract integrals

```python
# area between curves

lower = xPoints[0]
upper = xPoints[1]

top = sympy.integrate(eq1, (x, lower, upper))
bottom = sympy.integrate(eq2, (x, lower, upper))

area = abs(top - bottom)
print("Area = ", area)
```
