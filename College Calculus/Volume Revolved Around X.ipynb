# Section 1 : Good 3d graph

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d

# x range
lower = 0
upper = 3

# set up the graph
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
points = 10*(upper - lower)
u = np.linspace(lower, upper, points)
v = np.linspace(0, 2*np.pi, 60)
U, V = np.meshgrid(u, v)
X = U

# define function with capital U
fun = U**2
Y1 = (fun)*np.cos(V)
Z1 = (fun)*np.sin(V)

ax.plot_surface(X, Y1, Z1, alpha=0.4, color='blue')
plt.show()
```

# Section 2 : calculate volume

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# function defining the radius
radius = x**2

# pi * r-squared
area = sympy.pi * radius**2

lower = 0
upper = 3
vol = sympy.integrate(area, (x, lower, upper))
vol
```

```python
ar = sympy.N(vol)
ar 
```

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# same bounds for both functions 
lower = 0
upper = 3

# outer volume
outerRadius = x**2 + 2
outerArea = sympy.pi * outerRadius ** 2
outerVolume = sympy.integrate(outerArea, (x, lower, upper))

# inner volume
innerRadius = 2
innerArea = sympy.pi * innerRadius ** 2
innerVolume = sympy.integrate(innerArea, (x, lower, upper))

# difference
volume = outerVolume - innerVolume
volume
```

```python
ar = sympy.N(volume)
ar 
```
