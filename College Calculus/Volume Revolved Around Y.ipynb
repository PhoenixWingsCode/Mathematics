# Section 3 : Hollow Shape Volume

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d

# x range
lower = 0
upper = 3

# set up the graph
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
points = 10 * (upper - lower)
u = np.linspace(lower, upper, points)
v = np.linspace(0, 2*np.pi, 60)
U, V = np.meshgrid(u, v)
Z = U

# define function with capital U
fun = U**2 + 2
X1 = (fun)*np.cos(V)
Y1 = (fun)*np.sin(V)

fun2 = 1
X2 = (fun2)*np.cos(V)
Y2 = (fun2)*np.sin(V)

ax.plot_surface(X1, Y1, Z, alpha=0.3, color='red')
ax.plot_surface(X2, Y2, Z, alpha=0.2, color='blue')
plt.show()
```

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# same bounds for both functions 
lower = 0
upper = 3

# outer volume
outerRadius = y**2 + 2
outerArea = sympy.pi * outerRadius ** 2
outerVolume = sympy.integrate(outerArea, (y, lower, upper))

# inner volume
innerRadius = 2
innerArea = sympy.pi * innerRadius ** 2
innerVolume = sympy.integrate(innerArea, (y, lower, upper))

# difference
volume = outerVolume - innerVolume
volume
```

```python
ar = sympy.N(volume)
ar 
```
