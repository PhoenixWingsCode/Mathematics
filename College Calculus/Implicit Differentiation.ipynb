When "y" is not always on the other side of the equal sign

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# equation equal to zero
eq = x**2 + y**2 - 25

derivative = sympy.idiff(eq, y, x) # because y is a function of x
eq
```

```python
derivative
```

```python
from sympy.plotting import plot

sympy.plot_implicit(eq)
```

```python
# need an (x, y) point on the original line
xValue = 3
yValue = 4

# two steps to plug in x and y
answerX = derivative.subs(x, xValue)
answerY = answerX.subs(y, yValue)

# slope at (x,y)
print(answerY)
```

Better for multiple values

```python
# start with an x value:
xValue = 3

# find the y value in the original equation
xPart = eq.subs(x, xValue)
yPart = sympy.Eq(xPart, 0)
yValue = sympy.solve(yPart, y)
print("Y = ", yValue)

# you may get multiple y values
```

```python
for yLoop in yValue:
    # two steps to plug x and y into the derivative
    nswerX = derivative.subs(x, xValue)
    answerY = answerX.subs(y, yLoop)
    print("Slope at (", xValue, ",", yLoop, ") is", answerY)
```
