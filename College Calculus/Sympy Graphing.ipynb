# Section 1 : Simple Graph

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# original function
eq = x**2 - 3
sympy.plot(eq)
```

```python
sympy.plot(eq, title=None, xlabel=None, ylabel=None, zlabel=None, aspect_ratio='auto', xlim=None, ylim=None, axis_center='auto', axis=True, xscale='linear', yscale='linear', legend=False) # and many more
# first argument equation need that rest are optional

# Use options to modify the graph
```

# Section 3 : Shading

```python
import sympy
import numpy as np

x = sympy.symbols('x')

f = 1/(sympy.sqrt(2*sympy.pi))*sympy.exp(-(x**2)/2)
criticalX = 1.64
xArray = np.linspace(-4, criticalX, 1000)
fArray = sympy.lambdify(x, f)(xArray)
sympy.plot(f, (x, -4, 4), fill={'x':xArray,'y1':fArray,'color':'blue'})
```
