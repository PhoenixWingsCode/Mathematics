# Section 1 : Derivatives

Display the y = equation

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# Put the equation here:
eq = x**3 - 9*x**2 + 24
print("y = ")
eq
```

# Section 2 : Graph the function

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq) # ylim = [min, max]
```

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq, ylim = [-100, 100]) # ylim = [min, max]
```

# Section 3 : Display the derivative

```python
derivative = sympy.diff(eq, x)
print("Derivative = ")
derivative
```

# Section 4 : Find the slope(derivative) at a point

```python
xValue = 2
answer = derivative.subs(x, xValue)
print("slope at x = ", xValue, " is ", answer)
```

# Section 5 : Set derivative equal to zero

```python
first = sympy.Eq(derivative, 0)
first
```

# Section 6 : Solve to find critical values

```python
zeros = sympy.solve(first)
zeros
```

# Section 7 : Graph the original function and its derivative

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq, derivative)
```

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq, derivative, ylim=[-100,100])
```

# Section 8 : Second Derivative

Display the second derivative

```python
secondDerivative = sympy.diff(derivative, x)
secondDerivative
```

# Section 9 : Second Derivative at a point

```python
xValue = 6
answer = secondDerivative.subs(x, xValue)
print("slope at x = ", xValue, " is ", answer)
```

# Section 10 : Set second derivative equal to zero and solve

```python
second = sympy.Eq(secondDerivative, 0)
zeros = sympy.solve(second)
zeros
```

# Section 11 : Plot all the three functions

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq, derivative, secondDerivative, ylim=[-100,100])
```
