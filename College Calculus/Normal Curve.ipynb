# Section 1 : Normal Curve

y = 1/sqrt(2x)**e**-x**2/2

if the mean is 0 and the standard deviation = 1

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

eq = 1/(sympy.sqrt(2*sympy.pi))*sympy.exp(-(x**2)/2)
eq
```

```python
eq2 = eq*(sympy.sqrt(2)/sympy.sqrt(2))
eq2
```

# Section 2 : Plot

```python
sympy.plot(eq)
```

```python
sympy.plot(eq, xlim=[-4, 4])
```

```python
# area under the curve
area = sympy.integrate(eq, (x, -sympy.oo, sympy.oo))
area
```

# Section 3 : Area and Probability

```python
# partial area = probability
lower = -1
upper = 1
area = sympy.integrate(eq, (x, lower, upper))
print(sympy.N(area))
area
```

# Section 4 : Integral is the error function

```python
shaded = sympy.integrate(eq)
shaded
```

```python
sympy.plot(shaded, xlim=[-5, 5])
```

```python
# again, error function
#lower = shaded.subs(x, sympy.oo)
#lower
```

```python
# again, error function
lower = shaded.subs(x, -sympy.oo)
lower
```

```python
# Section 5 : Critical value to reject null hypothesis

step1 = sympy.Eq(shaded - lower, 0.95)
step2 = sympy.solve(step1)
step2
```

# Section 5 : More complex version

(1/sigma*(sqrt(2*x)))**e*-1/2(x-standard deviation/sigma)**2

```python
import sympy
from sympy import symbols

x, y = symbols('x y')
m, s = symbols('m s')

# adjust mean and standard deviation
m = 0
s = 1
denominator = s*sympy.sqrt(2*sympy.pi)
exponent1 = ((x - m)/s)**2
exponent2 = sympy.exp(-0.5*exponent1)
bell = 1/denominator*exponent2
bell
```

```python
# Plot

sympy.plot(bell, xlim=[m-4*s, m+4*s])
```

```python
import sympy

x = sympy.Symbol('x')
bell = sympy.exp(-x**2 / 2) / sympy.sqrt(2 * sympy.pi)
erfunction = sympy.integrate(bell)

# Check the coefficient
print(sympy.N(sympy.sqrt(2)*0.353553390593274))

# Area between values
lower = erfunction.subs(x, -sympy.oo)
upper = erfunction.subs(x, 0.682)
print("upper = ", sympy.N(upper))
print("lower = ", sympy.N(lower))
print("probability = ", sympy.N(upper - lower))
```
