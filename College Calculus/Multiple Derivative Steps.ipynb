# Section 1 : Derivatives

Display the y = equation

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# Put the equation here : 
eq = 3*x**2
print("y = ")
eq
```

# Section 2 : Display the derivative

```python
derivative = sympy.diff(eq, x)
print("Derivative = ")
derivative
```

# Section 3 : Find the slope (derivative) at a point

```python
xValue = 3
answer = derivative.subs(x, xValue)
print("Slope at x = ", xValue, " is ", answer)
```

# Section 4 : Set derivative equal to zero

```python
first = sympy.Eq(derivative, 0)
first
```

# Section 5 : Solve to find critical values

```python
zeros = sympy.solve(first)
zeros
```

# Section 6 : Graph the original function and its derivative

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq, derivative)
```

# Section 7 : Second Derivative

Display the second derivative

```python
secondDerivative = sympy.diff(derivative, x)
secondDerivative
```

# Section 8 : Second Derivative at a point

```python
xValue = 3
answer = secondDerivative.subs(x, xValue)
print("Slope at x = ", xValue, " is ", answer)
```

# Section 9 : Set second derivative equal to zero and solve

```python
second = sympy.Eq(secondDerivative, 0)
zeros = sympy.solve(second)
zeros
```

# Section 10 : Plot all three functions

```python
from sympy.plotting import plot

# plot the equation and its derivative
sympy.plot(eq, derivative, secondDerivative)
```

# Section 11 : Implicit Differentiation

When "y" is not always on the other side of the equal sign

```python
import sympy
from sympy import symbols

x, y = symbols('x y')

# equation equal to zero
eq = 3*x**2
derivative = sympy.idiff(eq, y, x)
eq
```

```python
# notice the plot_implicit
sympy.plot_implicit(eq)
```

```python
derivative
```

# Section 12 : You need an (x,y) point to find the slope

```python
# start with an xValue:
xValue = 3

# find the y value in the original equation
xPart = eq.subs(x, xValue)
yPart = sympy.Eq(xPart, 0)
yValue = sympy.solve(yPart, y)
print("Y values : ", yValue)

# you may get multiple y values
for yLoop in yValue:
    answerX = derivative.subs(x, xValue)
    answerY = answerX.subs(y, yLoop)
    print("slope at (", xValue, ",", yLoop, ") is ", answerY)
```

# Section 13 : Plot the equation along with the one point

```python
import sympy as sp

x, y = sp.symbols('x y')
eq = sp.Eq(x**2 + y**2, 4)  # example: circle of radius 2

xValue = 1
ySolutions = sp.solve(eq.subs(x, xValue), y)

if ySolutions:
    y1 = ySolutions[0]
    x1 = xValue
    sp.plot_implicit(eq, xlim=(-4, 4), ylim=(-4, 4), markers=[{'args': [x1, y1, 'ro']}])
else:
    print("No solution found for x =", xValue)
```
