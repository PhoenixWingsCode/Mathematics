```python
import sympy
from sympy import symbols, Eq

x, y, c = symbols('x y c')

# x value range
lower = 0
upper = sympy.pi

# for area between eq1 and x axis
eq1 = sympy.sin(x)

mass = sympy.integrate(eq1, (x, lower, upper))
eq2 = (eq1**2)/2
mx = sympy.integrate(eq2, (x, lower, upper))
eq3 = x*eq1
my = sympy.integrate(eq3, (x, lower, upper))
xCenter = my/mass
yCenter = mx/mass
print("x center of mass = ", xCenter)
print("y center of mass = ", yCenter)
print("point = (", sympy.N(xCenter), ",", sympy.N(yCenter),")")

# plot area with point in the center
sympy.plot(eq1, (x, lower, upper), markers=[{'args' : [xCenter, yCenter, 'ro']}])
```
