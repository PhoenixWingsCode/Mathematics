# Section 1 : Parametric derivatives

```python
import sympy
from sympy import symbols

x, y, t = symbols('x y t')

y = sympy.sin(t)
x = sympy.cos(t)
# no output
```

# Section 2 : Array graph

```python
import math
import matplotlib.pyplot as plt

xValues = []
yValues = []

for i in range(36):
    a = math.pi * i/12
    xt = x.subs(t, a)
    yt = y.subs(t, a)
    xValues.append(xt)
    yValues.append(yt)

# this plot works better for arrays
plt.plot(xValues, yValues)
plt.show()
```

# Section 3 : Sympy graph

```python
from sympy.plotting import plot_parametric
sympy.plot_parametric(x, y, (t, 0, 2 * sympy.pi))
```

# Section 4 : Separate derivatives, just because

```python
dydt = sympy.diff(y, t)
dydt
```

```python
dxdt = sympy.diff(x, t)
dxdt
```

```python
dydx = dydt/dxdt
dydx
```

# Section 5 : Slope at a specific t value

```python
import math

# math.pi or sympy.pi for different answers
slope = dydx.subs(t, math.pi/12)
print("Slope = ", slope)
slope = dydx.subs(t, sympy.pi/12)
slope
```
