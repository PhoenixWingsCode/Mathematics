# Section 4 : Bar Plots

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Plot
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Plot
plt.barh(categories, values, color='lightgreen')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [4, 7, 1, 8]
values2 = [5, 6, 2, 7]

x = np.arange(len(categories))  # X positions
width = 0.35  # Bar width

# Plot
plt.bar(x - width/2, values1, width, label='Group 1', color='orange')
plt.bar(x + width/2, values2, width, label='Group 2', color='blue')

plt.xticks(x, categories)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Plot')
plt.legend()
plt.show()
```

# Section 6 : Pie chart

```python
# Data for the pie chart
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 20, 20]  # Percentages
colors = ['red', 'yellow', 'pink', 'brown']
explode = (0.1, 0, 0, 0)  # Highlight the first slice (Apples)

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)

# Add a title
plt.title('Fruit Distribution')
```

# Section 8 : Plot customization

```python
years = [2014, 2015, 2016, 2017,
         2018, 2019, 2020, 2021]

income = [55, 56, 62, 61,
          72, 72, 73, 75]

incomeTicks = list(range(50, 81, 82))

plt.plot(years, income)
```

```python
plt.title("Income of John (in USD)")
plt.plot(years, income)
plt.xlabel("Year")
plt.ylabel("Yearly Income of USD")
```

```python
years = [2014, 2015, 2016, 2017,
         2018, 2019, 2020, 2021]

income = [55, 56, 62, 61,
          72, 72, 73, 75]

incomeTicks = list(range(50, 81, 2))

plt.title("Income of John (in USD)")
plt.plot(years, income)
plt.xlabel("Year")
plt.ylabel("Yearly Income of USD")
plt.yticks(incomeTicks, [f"{x}k USD" for x in incomeTicks])

plt.show()
```

```python
years = [2014, 2015, 2016, 2017,
         2018, 2019, 2020, 2021]

income = [55, 56, 62, 61,
          72, 72, 73, 75]

incomeTicks = list(range(50, 81, 2))

plt.title("Income of John (in USD)", fontsize=30, fontname="Agency FB")
plt.plot(years, income)
plt.xlabel("Year")
plt.ylabel("Yearly Income of USD")
plt.yticks(incomeTicks, [f"{x}k USD" for x in incomeTicks])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 101)  # avoid log(0) error

fig, axis = plt.subplots(2, 2, figsize=(10, 6))  # optional: set figure size

axis[0, 0].plot(x, np.sin(x))
axis[0, 0].set_title("Sine wave")

axis[0, 1].plot(x, np.cos(x))
axis[0, 1].set_title("Cosine wave")

axis[1, 0].plot(x, np.random.random(100))
axis[1, 0].set_title("Random Function")

axis[1, 1].plot(x, np.log(x))
axis[1, 1].set_title("Log function")
axis[1, 1].set_xlabel("Test")

fig.suptitle("Four Plots")

plt.tight_layout()
plt.show()
```

# Section 12 : 3d plotting 

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
axis = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

axis.scatter(x, y, z)
axis.set_title("3D Plot")

# Rotate the view
axis.view_init(elev=30, azim=45)  # adjust these values to rotate

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
axis = plt.axes(projection="3d")

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

axis.scatter(x, y, z)
axis.set_title("3D Plot")

# Rotate the view
axis.view_init(elev=30, azim=45)  # adjust these values to rotate

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
axis = plt.axes(projection="3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

axis.plot_surface(X, Y, Z)
axis.set_title("3D Plot")

# Rotate the view
axis.view_init(elev=30, azim=45)  # adjust these values to rotate

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
axis = plt.axes(projection="3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

axis.plot_surface(X, Y, Z, cmap="Spectral")
axis.set_title("3D Plot")

# Rotate the view
axis.view_init(elev=30, azim=45)  # adjust these values to rotate

plt.show()
```
