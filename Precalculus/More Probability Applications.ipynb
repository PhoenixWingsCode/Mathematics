# Section 2 : Normal distribution equation

y = e**(-x**2/2)/sqrt(2*x)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define plot window
xmin = -3
xmax = 3
ymin = -0.5
ymax = 1
points = 10 * (xmax - xmin)

# Generate x values
x = np.linspace(xmin, xmax, int(points))

# Standard normal distribution formula
y1 = (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

# Plot setup
fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'black')  # x-axis
plt.plot([0, 0], [ymin, ymax], 'black')  # y-axis
plt.plot(x, y1, label='Standard Normal Curve', color='blue')

plt.legend()
plt.title("Standard Normal Distribution")
plt.show()
```
