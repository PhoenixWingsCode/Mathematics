# Section 1 : Simple example

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path #path 'P' in capital letters
import matplotlib.patches as patches

verts = [] #An array of different vertices
steps = [] #Steps

#Add starting point
verts.append((0., 0.)) #an vertex of starting point making it float
steps.append(Path.MOVETO)

#Add lines one at a time
verts.append((2., 2.))
steps.append(Path.LINETO)

#Draw more lines as and when required

#graph
fig, ax = plt.subplots()
plt.axis((-5, 5, -5, 5))
path = Path(verts, steps)
patch = patches.PathPatch(path)
ax.add_patch(patch)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path #path 'P' in capital letters
import matplotlib.patches as patches

verts = [] #An array of different vertices
steps = [] #Steps

#Add starting point
verts.append((0., 0.)) #an vertex of starting point making it float
steps.append(Path.MOVETO)

#Add lines one at a time
verts.append((2., 2.))
steps.append(Path.LINETO)

#Draw more lines as and when required
verts.append((2., -2.))
steps.append(Path.LINETO)

verts.append((0., 0.))
steps.append(Path.CLOSEPOLY)

#graph
fig, ax = plt.subplots()
plt.axis((-5, 5, -5, 5))
path = Path(verts, steps)
patch = patches.PathPatch(path)
ax.add_patch(patch)
plt.show()
```

# Section 2 : More elaborate arrays

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches

#Array of points
verts = [(0., 0.), (-1., 1.), (0., 2.), (2., 2.), (3., 1.), (2., 0.), (0., 0.),]

#Array of steps
codes = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY]

#Graph
fig, ax = plt.subplots()
plt.axis((-5, 5, -5, 5))
path = Path(verts, codes)
patch = patches.PathPatch(path, facecolor='blue', lw=2) #Line width
ax.add_patch(patch)
plt.show() 
```
