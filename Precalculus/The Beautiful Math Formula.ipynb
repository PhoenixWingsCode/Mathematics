# Section 2 : Showing Robert Cotes was right in 1714

```python
import math
import cmath

# A way to represent
cNum = complex(0, 1)

for x in range(63):
    a = 0.1*x
    right = math.cos(a) + cNum*math.sin(a)
    left = math.e**(cNum*a)
    if left==right:
        print(round(a,1), "e value: ", left)
        print("cos, sin: ", math.cos(a), " + ", math.sin(a), "i")
```
