# Section 1 : Mandelbrot GPT code

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, maxIter):
    z = 0
    for i in range(maxIter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return maxIter

def plot_mandelbrot(xmin, xmax, ymin, ymax, width, height, maxIter):
    x = np.linspace(xmin, xmax, width)
    y = np.linspace(ymin, ymax, height)
    x, y = np.meshgrid(x, y)
    z = x + 1j * y
    img = np.zeros((height, width))

    for i in range(height):
        for j in range(width):
            c = z[i, j]
            img[i, j] = mandelbrot(c, maxIter)

    plt.imshow(img.T, cmap='hot', extent=[xmin, xmax, ymin, ymax])
    plt.xlabel('Re(c)')
    plt.ylabel('Im(c)')
    plt.title('Mandelbrot Set')
    plt.show()

# Define the parameters for the plot
xmin = 0.32
xmax = 0.36
ymin = -0.54
ymax = -0.50
width = 800
height = 600
maxIter = 100

# Call the plotting function
plot_mandelbrot(xmin, xmax, ymin, ymax, width, height, maxIter)
```
