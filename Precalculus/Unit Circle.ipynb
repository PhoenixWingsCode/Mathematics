# Section 1 : The Unit Circle

```python
import matplotlib.pyplot as plt
from IPython import display

fig, ax = plt.subplots()
plt.axis([-1.5,1.5,-1.5,1.5])
plt.plot([-2,2],[0,0],'b')
plt.plot([0,0],[-2,2],'b')
circle = plt.Circle((0,0), 1, fill = False) #(0,0) -> center 1 -> radius
ax.add_artist(circle)
ax.set_aspect('equal')

plt.show()
```

# Section 2 : Triangles in the Unit Circle

```python
import matplotlib.pyplot as plt
from IPython import display
import math

fig, ax = plt.subplots()
plt.axis([-1.5,1.5,-1.5,1.5])
plt.plot([-2,2],[0,0],'b')
plt.plot([0,0],[-2,2],'b')
circle = plt.Circle((0,0), 1, fill = False) #(0,0) -> center 1 -> radius
ax.add_artist(circle)
ax.set_aspect('equal')

# the triangle

angle = 1
x = math.cos(angle)
y = math.sin(angle)
plt.plot([x], [y], 'r')
plt.plot([0,x], [0,y], 'r')
plt.plot([0,x], [0,0], 'r')
plt.plot([x,x], [0,y], 'r')
plt.plot([x], [y], 'r')

ax.text(0.2, -0.1, 'x')
ax.text(0.6, 0.4, 'y')
ax.text(0.1, 0.6, 'r=1')
ax.text(x+0.1, y+0.1, '(x,y)')

plt.show()
```

# Section 3 : Animated Unit Circle

```python
import matplotlib.pyplot as plt
from IPython import display
import math
import time

#the triangles
for a in range(70):
    time.sleep(0.3)
    display.clear_output(wait=True)
    fig, ax = plt.subplots()
    circle = plt.Circle((0, 0), 1, fill=False)
    ax.set_aspect('equal')
    ax.add_artist(circle)
    plt.axis([-1.5,1.5,-1.5,1.5])
    plt.plot([-2,2],[0,0],'b')
    plt.plot([0,0],[-2,2],'b')

    angle = 0.1*a #a is counting variable when a is 1 it is 0.1 radian
    x = math.cos(angle)
    y = math.sin(angle)

    plt.plot([x], [y], 'r')
    plt.plot([0,x], [0,y], 'r')
    plt.plot([0,x], [0,0], 'r')
    plt.plot([x,x], [0,y], 'r')

    plt.show()
```

# Section 4 : Pythagorean trig

sin**2*x + cos**2*x = 1

```python
import math

for a in range(70):
    angle = 0.1*a
    y = math.sin(angle)**2
    x = math.cos(angle)**2
    r = x + y
    print(x, " + ", y, " = ", r)
```

# Section 7 : Animated Unit circle (Snapping Clock)

```python
import matplotlib.pyplot as plt
from IPython import display
import math
import time

#the triangles
for a in range(70):
    time.sleep(0.3)
    display.clear_output(wait=True)
    fig, ax = plt.subplots()
    circle = plt.Circle((0, 0), 1, fill=False)
    ax.set_aspect('equal')
    ax.add_artist(circle)
    plt.axis([-1.5,1.5,-1.5,1.5])
    plt.plot([-2,2],[0,0],'b')
    plt.plot([0,0],[-2,2],'b')

    angle = -0.1*a + math.pi/2 #a is counting variable when a is 1 it is 0.1 radian
    x = math.cos(angle)
    y = math.sin(angle)

    #plt.plot([x], [y], 'r')
    plt.plot([0,x], [0,y], 'r')
    #plt.plot([0,x], [0,0], 'r')
    #plt.plot([x,x], [0,y], 'r')

    plt.show()
```
