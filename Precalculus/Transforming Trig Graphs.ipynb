# Section 1 : Sine wave transformations

```python
import matplotlib.pyplot as plt
from IPython import display
import numpy as np
import math
%matplotlib inline
from ipywidgets import interactive

#Graph dimensions
xmin = -2
xmax = 14
ymin = -5
ymax = 5
points = 10*(xmax - xmin)
x = np.linspace(xmin, xmax, points)

def graph(a, b, c):
    fig, ax = plt.subplots()
    eq = 'sin('
    eq2 = 'x) + '
    plt.title(f'{a}{eq}{b}{eq2}{c}')
    plt.axis([xmin, xmax, ymin, ymax])
    plt.plot([xmin, xmax], [0, 0], 'b')
    plt.plot([0, 0], [ymin, ymax], 'b')
    y = a*np.sin(b*x)+c
    plt.plot(x, y, 'r')
    plt.show()

ig = interactive(graph, a=(-5,5), b=(-5,5), c=(-5,5))
ig
```

# Section 2 : Interacting sine waves

```python
# Notice the patterns when you add two sine functions ("additive interference")
# This is how noise cancelling headphones work and there is enough time to do all of this and still process the audio such that
# it detects the waves and then the generate the exact opposite so that what you hear is zeroed out sound

import matplotlib.pyplot as plt
from IPython import display
import numpy as np
import math

# Graph dimensions
xmin = -2
xmax = 14
ymin = -2
ymax = 2
points = 10*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'k')
plt.plot([0, 0], [ymin, ymax], 'k')

# change these
y1 = np.sin(x)
y2 = np.sin(x-math.pi)
y3 = y1 + y2

plt.plot(x, y1, 'b')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'r')
plt.show()
```
