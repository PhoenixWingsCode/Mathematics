# Section 1 : Just hours and minutes

```python
import matplotlib.pyplot as plt
from IPython import display
import math
import time
from datetime import datetime

# Set up plot and clock face
fig, ax = plt.subplots()
circle = plt.Circle((0, 0), 1, fill=False) #center, radius, fill for colors
ax.set_aspect('equal') #because python skewes up x and y
ax.add_artist(circle)
plt.axis([-1.5, 1.5, -1.5, 1.5])

# Dots to represent the 12 numbers
for b in range(13):
    plt.plot([math.cos(b*math.pi/6)],[math.sin(b*math.pi/6)],'b^') # x, y, blue triangle

# Minutes
timeNow = datetime.now()
min = int(timeNow.strftime("%M"))
minAngle = -min*math.pi/30 + math.pi/2 #2*Pie/60
plt.plot([0, 0.9*math.cos(minAngle)], [0, 0.9*math.sin(minAngle)], 'b')

# Hours
hours = int(timeNow.strftime("%H")) # No -8 or any minus here as it works perfectly
hourAngle = -hours*math.pi/6 + math.pi/2
plt.plot([0, 0.7*math.cos(hourAngle)], [0, 0.7*math.sin(hourAngle)], 'b')
plt.show()
```

# Section 2 : Loop for seconds - all hands in the loop

```python
import matplotlib.pyplot as plt
from IPython import display
import math
import time
from datetime import datetime

for a in range(70):
    time.sleep(1)
    display.clear_output(wait=True)
    
    # Setup figure and axis
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    
    # Clock border
    circle = plt.Circle((0, 0), 1, fill=False)
    ax.add_artist(circle)

    # Hour markers
    for b in range(12):
        angle = b * math.pi / 6
        ax.plot([math.cos(angle)], [math.sin(angle)], 'b^')

    # Current time
    timeNow = datetime.now()
    seconds = timeNow.second
    minutes = timeNow.minute
    hours = timeNow.hour % 12 + minutes / 60  # fractional hour

    # Second hand
    sec_angle = -seconds * math.pi / 30 + math.pi / 2
    ax.plot([0, math.cos(sec_angle)], [0, math.sin(sec_angle)], 'r')

    # Minute hand
    min_angle = -minutes * math.pi / 30 + math.pi / 2
    ax.plot([0, 0.9 * math.cos(min_angle)], [0, 0.9 * math.sin(min_angle)], 'b')

    # Hour hand
    hour_angle = -hours * math.pi / 6 + math.pi / 2
    ax.plot([0, 0.7 * math.cos(hour_angle)], [0, 0.7 * math.sin(hour_angle)], 'b')

    plt.show()
```
